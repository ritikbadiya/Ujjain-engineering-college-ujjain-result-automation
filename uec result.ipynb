```python
from selenium import webdriver     
import time 
import io
from selenium.webdriver.common.keys import Keys 
from PIL import Image
import numpy as np

import pytesseract
import sys
import argparse
from PIL import Image
from subprocess import check_output
import numpy as np
import cv2





def img_text(img):
    
    img=np.array(img)
    kernel = np.ones((3,3),np.uint8)

    dilate = cv2.dilate(img,kernel,iterations = 1)
    img_fi=Image.fromarray(dilate)
    return(''.join([i for i in list(resolve(img_fi).strip()) if i.isalnum()]))
# to solve capcha
def resolve(img):
    return pytesseract.image_to_string(img)   

#connect to installed web driver (chrome)

browser = webdriver.Chrome('./chromedriver_linux64/chromedriver') 
browser.get('http://www.uecu.ac.in/Results/BERslt.aspx')
path='/html/body/form/table/tbody/tr[2]/td/table/tbody/tr/td/table/tbody/tr/td/div/table/tbody/tr[1]/td/table/tbody/tr[2]/td/table/tbody/tr[3]/td/table/tbody/tr/td[2]/label'
ele1 = browser.find_elements_by_xpath(path)
ele1[0].click()
img_path='/html/body/form/table/tbody/tr[2]/td/table/tbody/tr/td/table/tbody/tr/td/div/table/tbody/tr[1]/td/table/tbody/tr[2]/td/div/table/tbody/tr[1]/td/div/img'



#from roll no 50 to 54

i=50


while i<55:
    try:
        browser.get('http://www.uecu.ac.in/Results/BERslt.aspx')
        path_en_no='/html/body/form/table/tbody/tr[2]/td/table/tbody/tr/td/table/tbody/tr/td/div/table/tbody/tr[1]/td/table/tbody/tr[2]/td/table/tbody/tr[2]/td[2]/input'
        en = browser.find_elements_by_xpath(path_en_no)
        en[0].send_keys('0701cs2010{}'.format(i))
        img= browser.find_elements_by_xpath(img_path)


        img = Image.open(io.BytesIO(img[0].screenshot_as_png))

        text=img_text(img)
        text_path='/html/body/form/table/tbody/tr[2]/td/table/tbody/tr/td/table/tbody/tr/td/div/table/tbody/tr[1]/td/table/tbody/tr[2]/td/div/table/tbody/tr[3]/td/input'
        box = browser.find_elements_by_xpath(text_path)
        box[0].send_keys(text)
        search_path='/html/body/form/table/tbody/tr[2]/td/table/tbody/tr/td/table/tbody/tr/td/div/table/tbody/tr[1]/td/table/tbody/tr[2]/td/div/table/tbody/tr[4]/td/input'
        search=browser.find_elements_by_xpath(search_path)
        search[0].click()
        i+=1
        time.sleep(5)
    except:
        i-=1
   
        

```
